Adding new column

```python
df = pd.read_csv('student_data (1).csv')
df
```

```python
#adding a column with a contrains values 

df['Country'] = 'Bangladesh'
df
```

```python
df['Total Marks'] = df['Data Structure Marks'] + df['Algorithm Marks'] + df['Python Marks']
df
```

```python
import numpy as np
df['A+ in Ds'] = np.where(df['Data Structure Marks'] > 80, "A+", "A")

df
```

```python
#pythn marks ar upore A+ dekhbo 
df["A+ for python"] = np.where(df["Python Marks"] > 80, "A+","A")
df
```

```python
df['First Name'] = df["FullName"].str.split(' ').str[0]
df
```

```python
#chaile ei noton dataframe ke ekta csv file a stroe kora jay 

df.to_csv('new_dataframe.csv')
```
