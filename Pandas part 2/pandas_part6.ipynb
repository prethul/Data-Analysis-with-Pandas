```python
import pandas as pd
```

```python
df = pd.read_csv('student_completed_data (1).csv')
df
```

```python
df['EnrollmentDate']
```

```python
df['EnrollmentDate'] = pd.to_datetime(df['EnrollmentDate'])
```

```python
df['EnrollmentDate']
```

```python
df['Enrollment Year'] = df['EnrollmentDate'].dt.year  
df
```

```python
df['Enrollment Date'] = df['EnrollmentDate'].dt.day
df
```

```python
df['FinishedDate'] = pd.to_datetime(df['FinishedDate'])
df['FinishedDate']
```

```python
df['Time taken to finish'] = df['FinishedDate'] - df['EnrollmentDate']
```

```python
df
```
