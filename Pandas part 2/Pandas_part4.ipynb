# Handleing null values

```python
import pandas as pd
```

```python
df = pd.read_csv('student_data (1).csv')
df
```

```python
#joto null values ache sob row coulmn delete hoye jabe 
df.dropna()
```

```python
#jekono ekta coulmn dore delete koraaaa 

df.dropna(subset=['Python Marks'])
```

```python
df.dropna(subset=['Python Marks','Algorithm Marks'])
```

```python
df.fillna(0)
```

```python
df['Python Marks'].fillna(60,inplace=True)
df
```

```python
df['Python Marks'].fillna(df['Python Marks'].mean(),inplace=True)
df
```

```python
df['Algorithm Marks'].fillna(df['Algorithm Marks'].mean(),inplace=True)
df
```

```python
df["Data Structure Marks"].fillna(df['Data Structure Marks'].mean(),inplace=True)
df
```
